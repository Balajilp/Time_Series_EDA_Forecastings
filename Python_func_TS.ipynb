```python
import datetime
import pandas as pd
import requests
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = (10, 8)
mpl.rcParams['axes.grid'] = False
```

```python
# fetching the datafrom the coindestk using request
coindeskURL = "http://api.coindesk.com/v1/bpi/historical/close.json?"

start = datetime.date(2017, 1, 1)
end = datetime.date(2020, 7, 2)

url = f'{coindeskURL}start={start:%Y-%m-%d}&end={end:%Y-%m-%d}'

result = requests.get(url)
result.content
```

```python
# loading the data
data = pd.read_json(result.content).iloc[:-2, :1]
data.index.name = 'date'
data.index = pd.to_datetime(data.index)
data
```

```python
data.plot()
```

```python
data.resample('Q').mean()
```

```python
data.resample('Q').mean().plot()
```

```python
data.plot()
data.resample('M').mean().plot()
```

## diff function
using diff function we can de trend our data and make it to stationary, so that we can use it on simple arima models


```python
data[:5]
```

```python
data[:5].diff()
```

```python
data[:5].diff(2)
```

```python
pd.concat([data['bpi'], data['bpi'].diff(2), data['bpi'].diff()], axis=1).plot()
```

## Shift function

```python
data[:5]
```

```python
data[:5].shift()
# here we are shifted the data to one step
```

```python
# now TSHIFT fucntion
# using this function we can shift the index date also
data[:5].tshift(2)
# above and below we can see the difference date
```

```python
pd.concat([data, data-data.shift(), data.tshift(365)], axis=1).plot()
# here green line is one year completely shifted
# blue is original line
# orange is detrended original blue line
```

## rolling function


```python
data.rolling(window=10).mean().plot()
```

```python
data.resample('w').mean().rolling(window=10).mean().plot()
```

```python
data.rolling(window=50).mean().plot()
data.rolling(window=50, win_type='gaussian').sum(std=10).plot()
```

```python
data.ewm(span=50).mean().plot()
```

```python
pd.concat([data, data.rolling(window=100).mean(), data.ewm(span=100).mean()], axis=1).plot()
```
